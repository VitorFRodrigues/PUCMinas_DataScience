### Gráfico de Dispersão
***
O gráfico de dispersão, também conhecido como scatter plot, é uma ferramenta poderosa para visualizar e analisar a relação entre duas variáveis. Ele permite identificar padrões, tendências e anomalias nos dados, ajudando a entender a natureza da relação entre as variáveis em estudo. O Matplotlib oferece recursos flexíveis para criar gráficos de dispersão, tornando a exploração de dados mais intuitiva e informativa.

Existem diversas situações em que o gráfico de dispersão pode ser utilizado para extrair insights valiosos.


<input type="checkbox" checked disabled> **Análise de correlação:** O gráfico de dispersão é especialmente útil para investigar a relação entre duas variáveis e identificar se existe uma correlação entre elas. Podemos analisar se há uma tendência de crescimento ou decrescimento conjunto, se os pontos estão dispersos aleatoriamente ou se há uma relação não linear.

<input type="checkbox" checked disabled> **Identificação de outliers:** Ao plotar os dados em um gráfico de dispersão, é possível visualizar pontos que se desviam significativamente do padrão estabelecido. Esses pontos são conhecidos como outliers e podem fornecer informações valiosas sobre dados incomuns, erros de medição ou eventos excepcionais.

<input type="checkbox" checked disabled> **Comparação de grupos:** O gráfico de dispersão também é útil para comparar grupos diferentes. Podemos atribuir diferentes cores ou formas aos pontos para representar diferentes categorias ou grupos de dados. Isso permite identificar padrões distintos entre os grupos e explorar as diferenças nas relações entre as variáveis.

<input type="checkbox" checked disabled> **Previsão e modelagem:** O gráfico de dispersão pode auxiliar na identificação de tendências ou padrões que podem ser úteis na construção de modelos preditivos. Ao visualizar a relação entre variáveis, podemos identificar comportamentos que podem ser incorporados em modelos de regressão ou aprendizado de máquina.

<input type="checkbox" checked disabled> **Análise de séries temporais:** Quando se trata de dados ao longo do tempo, o gráfico de dispersão pode ser utilizado para visualizar a evolução de uma variável em relação a outra. Isso pode revelar sazonalidades, tendências sazonais ou padrões cíclicos, permitindo uma compreensão mais aprofundada das relações temporais dos dados.




```python
sns.scatterplot(data=df, x='peso', y='colesterol')
plt.grid()
```

### Exemplo de um gráfico de dispersão para visualizar a relação entre duas variáveis 
***
Nesse exemplo vamos criar duas variáveis. Peso e altura e comparar a relação desses dados utilizado o gráfico de dispersão.

### Personalização de cores e marcadores nos gráficos de dispersão
***
No gráfico de dispersão também podemos escolher as cores e formas dos marcadores. 

```python
sns.scatterplot(data=df, x='peso', y='colesterol', hue='genero', style='genero', markers=['o', '^'])
plt.grid()
```

###  Exemplo prático 
***
Vamos imaginar que estamos realizando uma análise de um grupo de pessoas, coletando informações sobre altura, peso e idade. Nosso objetivo é investigar se existe alguma relação entre essas variáveis e entender melhor como elas se relacionam entre si.

Para isso, coletamos dados de 10 indivíduos e armazenamos em um DataFrame. Cada linha do DataFrame representa uma pessoa, e as colunas representam as variáveis: altura, peso e idade.

### Gráfico de Histograma
***
O histograma é uma das ferramentas mais utilizadas na análise estatística e na visualização de dados. Ele permite compreender a distribuição dos dados e identificar padrões, tendências e características importantes de um conjunto de observações. O Matplotlib oferece recursos flexíveis para criar histogramas, tornando a análise e a interpretação dos dados mais acessíveis.

O histograma representa graficamente a frequência de ocorrência de diferentes valores em um conjunto de dados. Ele divide o intervalo dos valores em uma série de intervalos (bins) e conta quantas observações se enquadram em cada intervalo. A altura das barras no histograma representa a frequência ou a proporção dessas observações, proporcionando uma visão visual da distribuição dos dados.

<input type="checkbox" checked disabled> **Análise da distribuição:** O histograma é útil para explorar a distribuição dos dados e identificar padrões como simetria, assimetria, curtose (achatamento) e presença de outliers. Ele pode revelar se os dados seguem uma distribuição normal, uniforme, exponencial, entre outras.

<input type="checkbox" checked disabled> **Identificação de intervalos críticos:** O histograma permite identificar intervalos nos quais os dados estão concentrados ou dispersos. Isso pode ser útil para identificar valores atípicos, lacunas nos dados ou agrupamentos específicos.

<input type="checkbox" checked disabled> **Detecção de assimetria:** O histograma pode mostrar se a distribuição dos dados é simétrica ou assimétrica. Uma distribuição assimétrica indica que os dados tendem a se acumular em uma extremidade do intervalo, podendo revelar informações importantes sobre o fenômeno ou variável em estudo.

<input type="checkbox" checked disabled> **Análise de tendências e padrões temporais:** Quando aplicado a dados temporais, o histograma pode fornecer informações sobre padrões de sazonalidade, flutuações ou mudanças ao longo do tempo. Isso é útil para análise de séries temporais, previsões e tomada de decisões baseadas em tendências.

<input type="checkbox" checked disabled> **Comparação de grupos:** É possível criar histogramas separados para diferentes grupos de dados e comparar suas distribuições. Isso permite identificar diferenças ou semelhanças entre grupos e explorar possíveis relações ou características distintas.

```python
sns.histplot(data=df, x='peso', bins=50, hue='genero',palette='hls')
```
